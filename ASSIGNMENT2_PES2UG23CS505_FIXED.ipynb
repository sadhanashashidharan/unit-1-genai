SOCIAL MEDIA ANALYSIS

Title- Social Media Sentiment & Brand Analysis using Hugging Face Pipelines

Problem Statement- Social media contains a large amount of user-generated content where people express opinions about brands and products. Manually analyzing this data is difficult and time-consuming. The objective of this project is to automatically identify brand mentions and analyze the sentiment of social media posts using NLP techniques.

METHODOLOGY

Input Collection Social media posts or comments are taken as text input.

Sentiment Analysis The text is analyzed using a Twitter-trained RoBERTa model to classify sentiment as Positive, Negative, or Neutral.

Named Entity Recognition (NER) Brand or organization names are extracted from the text using a NER model.

Result Display The detected entities and sentiment results are displayed as output.

```python
from transformers import pipeline

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-roberta-base-sentiment"
)

text = "I love the camera quality of the new iPhone but the battery life is terrible 😡"

result = sentiment_analyzer(text)
print(result)
```

Observation

The sentiment model returned the label LABEL_0 with a confidence score, indicating a negative sentiment.

Reason

The Twitter RoBERTa sentiment model outputs numerical labels which correspond to sentiment classes and require manual label mapping.
